```python
menu = {
    "Coffee" : {
        "Espresso" : 1.50,
        "Americano" : 2.00, 
        "Caffe latte" : 2.25,
        "Mocha" : 2.25,
        "Cappucino" : 2.50,
        "Macchiato" : 2.50
    },
    "Tea" : {
        "White tea" : 1.50,
        "Green tea" : 2.00,
        "Oolong tea" : 2.25,
        "Black tea" : 2.25,
        "Earl Grey" : 2.50,
        "Matcha" : 2.50
    },
    "Pastries" : {
        "Croissant" : 5.50,
        "Cinnamon roll" : 4.50,
        "Banana bread" : 4.25,
        "Pumpkin bread" : 3.75,
        "Sticky bun" : 4.25,
        "Apple turnover" : 3.75
    }
}

item_name = ""
ordered = {}
all_orders = {}

print("Welcome to MC Cafe!")

menu_answer = input("Do you want to see the menu? (y/n)")

if menu_answer == "y":
    for category_name, values in menu.items():
        print("\n" + category_name)
        for item, price in values.items():
            print("\t" + item + " :", price)

while True:
    item_answer = input("Do you want to order an item? (y/n)")
    if item_answer == "y":
        item_name = input("What would you like?")
        item_quantity = int(input(f"How many of {item_name.title()} would you like?"))
        ordered[item_name] = item_quantity
    else :
        break

#this function loops through the various menu items and returns the price 
#for a menu item
def find_price(item_name, products_available):
    item_price = 0
    for values in products_available.values():
        for item, price in values.items():
            if item_name.lower() == item.lower():
                item_price = price
    return item_price
#end of function

for item, qty in ordered.items():
    price = find_price(item, menu)
    all_orders[item] = {"qty" : qty, "price" : price}

total = 0

print("\nReceipt: ")
for item, value in all_orders.items():
    qty = value["qty"]
    price = value["price"]
    sub = qty * price
    total += sub
    print(f"{item.title()} \t {qty} @ {price:,.2f}")

tax = total * 0.06
print(f"\n\nSubtotal: \t{total:,.2f}")
print(f"Tax: \t\t{tax:,.2f}")
print(f"\nTotal: \t\t{total+tax:,.2f}")
```
