```python
menu = {
    "Espresso": 1.50,
    "Americano": 2.00,
    "Caffe latte": 2.25,
    "Mocha": 2.25,
    "Cappucino": 2.50,
    "Macchiato": 2.50,
    "White tea": 1.50,
    "Green tea": 2.00,
    "Oolong tea": 2.25,
    "Black tea": 2.25,
    "Earl Grey": 2.50,
    "Matcha": 2.50,
    "Croissant": 5.50,
    "Cinnamon roll": 4.50,
    "Banana bread": 4.25,
    "Pumpkin bread": 3.75,
    "Sticky bun": 4.25,
    "Apple turnover": 3.75
}

print("Welcome to MC Cafe!")
menuC = input("Would you like to see the menu? (y/n) ")

if menuC.lower() == "y":
  print("\nMenu item - Price")
  print("-"*20)
  for a, b in menu.items():
    print(f"{a} - ${b:.2f}")
  print()

order = {}
total = 0

orderC = input("Order? (y/n) ")
while orderC.lower() == "y":
    item = input("Enter item: ")
    qu = int(input("Enter quantity: "))
  
    if item in menu:
        sub = menu[item] * qu
      
        if item in order:
            order[item] += qu
        else:
            order[item] = qu
            total += sub   
    else:
        print(f"Try again, {item} is not on the menu.")
    orderC = input("Order again? (y/n) ")
    
print("\nOrder Receipt:")
print("-"*20)
for item, qu in order.items():
    print(f"{item}: {qu}")
print(f"\nSubtotal:    ${total:.2f}")
tax = total * 0.06
print(f"Tax:         ${tax:.2f}")
grand = total + tax
print(f"Grand Total: ${grand:.2f}\n")

print("Thank you for visiting MC Cafe!")
```
